```python
import yfinance as yf
```

```python
# Tesla'nın hisse senedi verilerini çekme
tesla_data = yf.Ticker("TSLA")
```

```python
# Hisse senedi fiyatları için veri çekme
tesla_stock_data = tesla_data.history(period="1d", interval="1m")
```

```python
# Finansal veriler için veri çekme (örneğin, gelir ve kar verileri)
tesla_financials = tesla_data.financials
```

```python
# Çekilen verileri gösterme
print("Tesla'nın Son 5 Hisse Senedi Fiyatı:")
print(tesla_stock_data.tail())
```

```python
print("\nTesla'nın Finansal Verileri:")
print(tesla_financials)
```
